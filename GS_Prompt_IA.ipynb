# Estrutura dos Prompts de Exemplo

**Antes, Durante e Depois do Desastre**


Para tornar o chatbot realmente útil em diferentes momentos de um desastre natural, os prompts de exemplo foram criados com base em três premissas principais:
Antes do acontecimento (Prevenção)
 Os prompts dessa fase simulam perguntas feitas por usuários em momentos de alerta ou risco iminente. O objetivo é preparar a pessoa com orientações preventivas, como:


“O que devo fazer se houver risco de enchente na minha região?”


“Como montar um kit de emergência?”


“É seguro continuar morando perto de uma encosta com rachaduras?”


Durante o acontecimento (Resposta Imediata)
 Nessa fase, os prompts são voltados para situações de emergência em tempo real. O chatbot foi treinado para oferecer respostas rápidas, objetivas e que priorizem a segurança, como:


“A água começou a entrar em casa, o que eu faço?”


“Estou ouvindo um estrondo vindo do morro, devo evacuar?”


“Estou em um local alto, mas sem comida nem água. O que devo priorizar?”


Depois do acontecimento (Recuperação e Cuidados)
 Após o desastre, o foco muda para cuidados com a saúde, segurança e reestruturação, com exemplos como:


“Posso voltar para casa depois da enchente?”


“Como evitar doenças após o contato com água contaminada?”


“Perdi meus documentos, onde posso buscar ajuda?”


Essa estrutura em três etapas permitiu que o chatbot oferecesse respostas adequadas ao contexto, respeitando o tempo de cada fase e mantendo o foco na proteção da vida e bem-estar da população afetada.


# Análise de Prompt:

**Prompt Usado:**
O que fazer se minha casa alagar?


**Resposta Gerada:**


1. **Fique Calmo**: Mantenha a calma e avalie a situação. É importante tomar decisões racionais durante momentos de estresse.

2. **Desconecte Energia**: Apague a eletricidade e desconecte os aparelhos elétricos, se for seguro fazê-lo. Isso ajuda a prevenir riscos de choque elétrico.

3. **Mova Itens Importantes**: Tente mover móveis, eletrodomésticos e outros itens valiosos para os andares mais altos da casa ou para áreas que não estão em risco de alagamento.

4. **Prepare um Kit de Emergência**: Tenha à mão um kit de emergência com itens essenciais como água, alimentos não perecíveis, medicamentos, lanternas, pilhas e documentos importantes.

5. **Siga as Orientações das Autoridades**: Ouça informações e recomendações da defesa civil local e de autoridades que informam sobre o estado do tempo e do alagamento.

6. **Evacue se Necessário**: Se as autoridades indicarem evacuação ou se as águas começarem a subir rapidamente, faça isso imediatamente. Procure locais seguros para se abrigar.

7. **Evite á e em Movimento**: Não tente dirigir através de estradas inundadas, pois a profundidade e a correnteza podem ser perigosas.

8. **Fique Em Contato**: Informe familiares ou amigos sobre sua situação, especialmente se precisar de ajuda.

Após a inundação, tome cuidado ao retornar para casa. Verifique se há danos e evite áreas onde possa haver água parada, que pode ser contaminada.

Priorize sempre sua segurança e a da sua família. Se precisar de ajuda, entre em contato com serviços de emergência ou organizações de apoio.

Prompt Usado: O que fazer se eu passar por um terremoto numa viagem que farei ao japão?

**Resposta Gerada:**

O que fazer se eu passar por um terremoto numa viagem que farei ao japão?
Se você passar por um terremoto enquanto estiver no Japão, é importante manter a calma e seguir algumas orientações de segurança:

1. **Se você estiver em um prédio**:
   - Procure abrigo sob uma mesa sólida ou entre os móveis.
   - Afaste-se de janelas, vidro e itens pesados que possam cair.
   - Fique onde está até que o tremor pare.

2. **Se você estiver fora**:
   - Vá para um espaço aberto, longe de prédios, postes e árvores.
   - Ajoelhe-se ou deite-se no chão e proteja a cabeça e o pescoço com os braços.

3. **Se você estiver dirigindo**:
   - Pare o carro em um local seguro, longe de prédios e fios elétricos.
   - Fique dentro do veículo até que o tremor pare e siga a orientação das autoridades.

4. **Após o tremor**:
   - Verifique se você e os outros ao seu redor estão seguros e sem ferimentos.
   - Esteja atento a possíveis réplicas, que podem ocorrer após o tremor inicial.
   - Siga as instruções das autoridades locais e mantenha-se informado sobre a situação.

5. **Tenha um plano de emergência**:
   - Antes da viagem, familiarize-se com o seu local, saiba onde ficam as saídas de emergência e identifique abrigos próximos.
   - Considere fazer um aplicativo de emergência do seu celular para receber informações em tempo real sobre desastres.

Lembre-se que manter a calma é essencial e pode ajudar você a tomar decisões seguras em uma situação de emergência. Boa viagem e cuide-se!
